# Assignment: Performance Metrics from Confusion Matrix

This notebook computes **Accuracy, Precision, Recall, and F1-score** from a given confusion matrix.

## Given Confusion Matrix

|              | Predicted Positive | Predicted Negative |
|--------------|-------------------|-------------------|
| **Actual Positive** | TP = 199 | FN = 1 |
| **Actual Negative** | FP = 3   | TN = 247 |

```python
# Confusion matrix values
TP = 199
FN = 1
FP = 3
TN = 247

TP, FN, FP, TN
```

## Metric Calculations

```python
# Accuracy
accuracy = (TP + TN) / (TP + TN + FP + FN)

# Precision
precision = TP / (TP + FP)

# Recall
recall = TP / (TP + FN)

# F1-Score
f1_score = 2 * (precision * recall) / (precision + recall)

accuracy, precision, recall, f1_score
```

## Final Results

- **Accuracy** = 0.9911
- **Precision** = 0.9851
- **Recall** = 0.9950
- **F1-score** = 0.9900

# Assignment 4: Z-Test for Mango Quality

A village has multiple mango gardens with an average of **85% good quality mangoes**. One person claims his garden produces **88% good quality mangoes**.

Assume:
- Gaussian distribution
- Standard deviation (σ) = 1

We use a **Z-score test** to accept or reject the hypothesis.

## Step 1: Hypotheses

- **Null Hypothesis (H₀):** μ = 85
- **Alternative Hypothesis (H₁):** μ = 88

## Step 2: Given Values

- Population mean (μ) = 85
- Sample mean (x̄) = 88
- Standard deviation (σ) = 1

```python
# Given values
mu = 85      # population mean
x = 88       # sample mean
sigma = 1    # standard deviation

mu, x, sigma
```

## Step 3: Z-Score Formula

Z = (x̄ − μ) / σ

```python
# Z-score calculation
z_score = (x - mu) / sigma
z_score
```

## Step 4: Decision Rule

At **95% confidence level**, critical Z-value = **±1.96**

```python
# Hypothesis decision
critical_value = 1.96

if abs(z_score) > critical_value:
    decision = 'Reject Null Hypothesis (Claim Accepted)'
else:
    decision = 'Accept Null Hypothesis (Claim Rejected)'

decision
```

## Final Conclusion

- Calculated Z-score = **3**
- Since |Z| > 1.96, the null hypothesis is rejected.

**Conclusion:** The person’s claim that his garden has **88% good quality mangoes is accepted.**
